## Data Preparation and Transformation

```python
# Importing all the libraries
import pandas as pd
```

```python
# Reading the data from the csv file
data = pd.read_excel(r'C:\Users\maker\Documents\ANL488\Gemini_BTCUSD_1h.xlsx')
```

```python
# Show the data
data.head()
```

```python
# Show the data types
data.dtypes
```

```python
# Calculating a 7-day simple moving average for the closing price
data['SMA_7'] = data['close'].rolling(window=7).mean()
# Calculating a 14-day simple moving average for the closing price
data['SMA_14'] = data['close'].rolling(window=14).mean()
data.head(20)
```

```python
# Calculating the 7-day exponential moving average for the closing price
data['EMA_7'] = data['close'].ewm(span=7, adjust=False).mean()
# Calculating the 14-day exponential moving average for the closing price
data['EMA_14'] = data['close'].ewm(span=14, adjust=False).mean()
data.head(20)
```

```python
# Calculating the MACD line
data['EMA_12'] = data['close'].ewm(span=12, adjust=False).mean()
data['EMA_26'] = data['close'].ewm(span=26, adjust=False).mean()
data['MACD'] = data['EMA_12'] - data['EMA_26']
#Calculating the signal line
data['Signal'] = data['MACD'].ewm(span=9, adjust=False).mean()
# Calculating the MACD histogram
data['Histogram'] = data['MACD'] - data['Signal']
data.head(20)
```

```python
# Calculating the RSI
delta = data['close'].diff()
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
rs = gain / loss
data['RSI'] = 100 - (100 / (1 + rs))
data.head(20)
```

```python
# Calculate the Bolllinger Bands
data['20 Day MA'] = data['close'].rolling(window=20).mean()
data['20 Day STD'] = data['close'].rolling(window=20).std()
data['Upper Band'] = data['20 Day MA'] + (data['20 Day STD'] * 2)
data['Lower Band'] = data['20 Day MA'] - (data['20 Day STD'] * 2)
data.head(20)
```

```python
#Calculate the ATR
data['H-L'] = abs(data['high'] - data['low'])
data['H-PC'] = abs(data['high'] - data['close'].shift(1))
data['L-PC'] = abs(data['low'] - data['close'].shift(1))
data['TR'] = data[['H-L', 'H-PC', 'L-PC']].max(axis=1)
data['ATR'] = data['TR'].rolling(window=14).mean()
data.head(20)
```

```python
# Account for price-invariance by changing the price to a percentage change
data['close_%'] = data['close'].pct_change()
data['high_%'] = data['high'].pct_change()
data['low_%'] = data['low'].pct_change()
data['open_%'] = data['open'].pct_change()
# Account for volume-invariance by changing the volume to a percentage change
data['volume_btc_%'] = data['Volume BTC'].pct_change()
data['volume_usd_%'] = data['Volume USD'].pct_change()

data.head(20)

```

```python
data.to_excel(r'Gemini_BTCUSD_1h.xlsx', index = False)
```
