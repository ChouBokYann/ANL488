```python
import pandas as pd
```

```python
merged_data = pd.read_csv('merged_data.csv')
```

```python
merged_data.head(5)
```

```python
# take out 2 weeks worth of data for testing
start_date = '2024-01-02'
end_date = '2024-01-16'

mask = (merged_data['date'] > start_date) & (merged_data['date'] <= end_date)
weeks2 = merged_data.loc[mask]

weeks2.head(5)
```

```python
# Find missing values in the data
missing_values = weeks2.isnull().sum()
missing_values
```

```python
weeks2.to_csv('2weeks.csv', index=False)
```
